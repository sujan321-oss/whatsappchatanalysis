# pattern to extract the time 


```python
time_pattern="\d{1,2}:\d{1,2}"
date_time_pattern="\d{1,2}/\d{1,2}/\d{1,2}, \d{1,2}:\d{1,2}"
```

```python
date_time=re.findall(date_time_pattern,file)
```

```python
len(date_time)
```

```python
df=pd.DataFrame({"date_time":date_time})
```

```python
date_list=[]
time_list=[]
def date_time_split(data):
    date=data.split(',')[0]
    time=data.split(',')[1]
    date_list.append(date)
    time_list.append(time)
        
    
    

```

```python
date_time=df["date_time"].apply(date_time_split)
```

```python
df['date']=date_list
```

```python
df['time']=time_list
```

```python
# extracting the name and the message 
name_pattern='(.+):'
```

```python
name_message=re.split(date_time_pattern,file)[1:]
```

```python
name_message
```

```python
name=[]
for i in name_message:
    match=re.search(name_pattern,i)
    if(match):
        name.append(re.findall(name_pattern,i)[0].split(":")[0])
    else:
        name.append("notification")
        
    
    
```

```python
name
```

```python
cleaned_name=[]
for i in name:
    cleaned_name.append(i.replace("-",""))
    
```

```python
cleaned_name
```

```python
len(cleaned_name)
```

```python
df['name']=cleaned_name
```
